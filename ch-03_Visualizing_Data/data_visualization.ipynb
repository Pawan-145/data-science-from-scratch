```python
from matplotlib import pyplot as plt  
```

```python
years = [1950,1960,1970,1980,1990,2000,2010]
gdp = [300.2,543.3,1075.9, 2862.5, 5979.6, 10289.7, 14958.3]

# create a line chart, years, on x-axis, gdp on y-axis
plt.plot(years,gdp, color='green', marker='o', linestyle='solid')

# add a title
plt.title("Nominal GDP")

# add a label to the y-axis
plt.ylabel("Billions of $")
plt.xlabel("Years")
plt.show()  
```

# Bar Charts

```python
movies = ["Avengers","Infinity War", "End Game","Avatar","Doomsday"]
num_oscars = [6,8,3,9,11]

# plot bars with left x-coordinates [0,1,2,3,4], heights[num_oscars]
plt.bar(range(len(movies)), num_oscars)
plt.title("My fav movies")
plt.ylabel("# of Academy Awards")

# label x-axis with movie names at bar centers 
plt.xticks(range(len(movies)),movies)
# plt.xlabel("Movies")
plt.show()
```

# Using barchat for a Histogram

```python
%pip install collections
```

```python
from collections import Counter

grades = [83,95,91,87,70,0,85,82,100,67,73,77,0]

# Bucket grades by decile, but put 100 in with the 90s
histogram = Counter(min(grade // 10*10,90) for grade in grades)

plt.bar([x+5 for x in histogram.keys()], histogram.values(),10,edgecolor=(0,0,0))
plt.axis([-5,105,0,5])               #x-axis from -5 to 105
                                     # y-axis from 0 to 5 

plt.xticks([10*i for i in range(11)])
plt.xlabel("Decile")
plt.ylabel("# of students")
plt.title("Distribution of Exam 1 Grades")
plt.show()
```

# Line Charts

```python
variance = [1,2,4,8,16,32,64,128,256]
bias_squared = [256,128,64,32,16,8,4,2,1]
total_error = [x+y for x,y in zip(variance,bias_squared)]
xs = [i for i,_ in enumerate(variance)]

#Alternate for xs
# xs = list(range(len(variance)))

# We can make multiple calls to plt.plot to show multiple series on the same chart
plt.plot(xs,variance, 'g-', label='variance')
plt.plot(xs,bias_squared, 'r-.', label='bias^2')
plt.plot(xs,total_error,'b:' , label='total error')

# Because we have assign label to each series, we can get a legend for free(loc=9 means 'top center')
plt.legend(loc=9)
plt.xlabel('model complexity')
plt.xticks([])
plt.title("The Bias-Variance Tradeoff")
plt.show()
```

# Scatterplots

```python
friends = [70, 65, 72, 63, 71, 64, 60 ,64, 67]
minutes = [175, 170, 205, 120, 220, 130, 105, 145, 190]
labels = ['a','b','c ','d','e','f','g','h','t']

plt.scatter(friends,minutes)

# label each point
for label, friend_count,minute_count in zip(labels,friends,minutes):
    plt.annotate(label,
                 xy=(friend_count,minute_count),xytext=(5,-5),textcoords='offset points')
    
plt.title("Daily Minutes vs Number of Friends")
plt.xlabel("# of friends")
plt.ylabel("daily minutes spent on the site")
plt.show()
```

```python
# If you are scatering comparable variables, you might get a misleading picture if you let matplolib choose the scale

test_1_grades = [99,90,85,97,80]
test_2_grades = [100,85,60,90,70]

plt.scatter(test_1_grades,test_2_grades)
plt.title("Axes aren't  comparable")
plt.xlabel("test_1_grades")
plt.ylabel("test_2_grades")
plt.show()
```

```python
# If we include a call to plt.axis("equal"), the plot more accurately show the most of the variation occur on test2
test_1_grades = [99,90,85,97,80]
test_2_grades = [100,85,60,90,70]
labels = ['a','b','c','d','e']

plt.scatter(test_1_grades,test_2_grades)
for label,test_1_grades,test_2_grades in zip(labels,test_1_grades,test_2_grades):
    plt.annotate(
        label, xy=(test_1_grades,test_2_grades), xytext=(5,-5), textcoords="offset points"
    )
plt.title("Axes aren't  comparable")
plt.axis("equal")
plt.xlabel("test_1_grades")
plt.ylabel("test_2_grades")
plt.show()
```
